2. Train test split và robust scaler cho các thuộc tính numerical

```python
processed_links = [r"ip-files-fillna-lstm\APACHE-MHDDoS-fillna.csv", 
                   r"ip-files-fillna-lstm\BYPASS-MHDDoS-1-fillna.csv",
                   r"ip-files-fillna-lstm\BYPASS-MHDDoS-2-fillna.csv",
                   r"ip-files-fillna-lstm\BYPASS-MHDDoS-fillna.csv",
                   r"ip-files-fillna-lstm\CFB-MHDDoS-fillna.csv",
                   r"ip-files-fillna-lstm\DGB-MHDDoS-fillna.csv",
                   r"ip-files-fillna-lstm\HEAD-MHDDoS-fillna.csv",
                   r"ip-files-fillna-lstm\HTTP-Flood-Xerox-fillna.csv",
                   r"ip-files-fillna-lstm\KILLER-MHDDoS-fillna.csv",
                   r"ip-files-fillna-lstm\Mixed-Get-Post-Head-fillna.csv",
                   r"ip-files-fillna-lstm\NULL-MHDDoS-fillna.csv",
                   r"ip-files-fillna-lstm\POST-Flood-MHDDoS-fillna.csv",
                   r"ip-files-fillna-lstm\STOMP-MHDDoS-fillna.csv",
]
```

```python
import os
import pandas as pd
import numpy as np
import gc

def reduce_mem_usage(df, filename=""):
    """
    Tự động quét qua các cột và ép kiểu dữ liệu xuống để tiết kiệm RAM.
    Đồng thời đưa các cột bị lỗi object/mixed types dứt khoát về dạng string.
    """
    start_mem = df.memory_usage().sum() / 1024**2
    
    for col in df.columns:
        col_type = df[col].dtype
        
        # Chỉ xử lý các cột dạng số (không phải chuỗi/object)
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            
            # Xử lý số Nguyên (Integer)
            if pd.api.types.is_integer_dtype(col_type):
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            
            # Xử lý số Thực (Float) - Ta ép về float32 là tốt nhất cho LSTM/CNN
            elif pd.api.types.is_float_dtype(col_type):
                if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        else:
            # Sửa Fix lỗi Parquet ArrowTypeError cho các cột object lẫn lộn
            # Ép ép dứt khoát tất cả object về kiểu chuỗi (string) để pyarrow không crash
            df[col] = df[col].astype(str)
                    
    end_mem = df.memory_usage().sum() / 1024**2
    if filename:
        print(f'[{filename}] Giảm bộ nhớ: {start_mem:.2f} MB -> {end_mem:.2f} MB (Tiết kiệm {100 * (start_mem - end_mem) / start_mem:.1f}%)')
    
    return df

# 1. Tạo thư mục lưu trữ nếu chưa tồn tại
os.makedirs("ip-files-train-base", exist_ok=True)
os.makedirs("ip-files-valid-base", exist_ok=True)
os.makedirs("ip-files-train-meta", exist_ok=True)
os.makedirs("ip-files-test", exist_ok=True)

print("-" * 50)
print("Bắt đầu quá trình chia Train/Test từng file...")

# 2. Xử lý từng file một
for file in processed_links:
    filename = file.split('\\')[-1].replace(".csv", ".parquet") # Đổi đuôi ra parquet cho chuẩn
    
    # 2.1. Đọc file (cố gắng giữ nguyên cấu trúc dòng chảy thời gian)
    df_temp = pd.read_csv(file, low_memory=False)
    
    # sắp xếp theo timestamp để đảm bảo tính liên tục thời gian, tránh rối loạn khi chia tập
    df_temp = df_temp.sort_values(by="timestamp").reset_index(drop=True)

    # 2.2. Ép kiểu giảm RAM và khắc phục lỗi Arrow (object to str)
    df_temp = reduce_mem_usage(df_temp, filename=filename)
    
    # 2.3. Chia ra 4 tập: train_base, valid_base, train_meta và test theo tỷ lệ (50, 15, 15, 20)
    total_len = len(df_temp)
    train_len = int(0.5 * total_len)
    train_valid_len = int(0.65 * total_len)
    train_meta_len = int(0.8 * total_len)
    
    df_train_base = df_temp.iloc[:train_len]
    df_valid_base = df_temp.iloc[train_len:train_valid_len]
    df_train_meta = df_temp.iloc[train_valid_len:train_meta_len]
    df_test = df_temp.iloc[train_meta_len:]

    # 2.4. Khai báo đường dẫn lưu
    train_base_path = os.path.join("ip-files-train-base", filename)
    valid_base_path  = os.path.join("ip-files-valid-base", filename)
    train_meta_path = os.path.join("ip-files-train-meta", filename)
    test_path = os.path.join("ip-files-test", filename)
    
    # Lưu ra dạng .parquet cực nhẹ, dữ liệu nén nhanh
    df_train_base.to_parquet(train_base_path, index=False)
    df_valid_base.to_parquet(valid_base_path, index=False)
    df_train_meta.to_parquet(train_meta_path, index=False)
    df_test.to_parquet(test_path, index=False)
    
    print(f'Đã lưu train_base vào {train_base_path}')
    print(f'Đã lưu valid_base vào {valid_base_path}')
    print(f'Đã lưu train_meta vào {train_meta_path}')
    print(f'Đã lưu test vào {test_path}')
    
    # 2.5. Giải phóng triệt để RAM
    del df_train_base, df_valid_base, df_train_meta, df_test, df_temp
    gc.collect()

print("Hoàn tất quá trình đọc, nén, chia tập và lưu file thành công!")
```

```python
from sklearn.preprocessing import QuantileTransformer
import glob
import os
import gc
import numpy as np

# 1. ĐỌC VÀ GỘP CÁC FILE TRONG FOLDER TRAIN
print("Bắt đầu đọc và gộp (merge) tập TRAIN...")
train_files = glob.glob(r'ip-files-train-base\*.parquet')
train_dfs = [pd.read_parquet(f) for f in train_files]
data_train_merged = pd.concat(train_dfs, ignore_index=True)

del train_dfs
gc.collect()
print(f"Tổng kích thước tập Train: {data_train_merged.shape}")

# XỬ LÝ LỖI INFINITY (VÔ CỰC):
# Trong dữ liệu mạng thường có các phép chia Float cho 0 sinh ra vô cực (Infinity / np.inf).
# Thuật toán ML của Sklearn sẽ bị sập nếu gặp Infinity. Chúng ta phải tìm và thay thế nó bằng giá trị Max (hoặc 0) hợp lý nhất.
# Tìm các cột số
numerical_cols = data_train_merged.select_dtypes(include=[np.number]).columns.tolist()

print("Đang dọn dẹp các giá trị Infinity (np.inf) trong tập Train...")
for col in numerical_cols:
    # Nếu cột có chứa cực đại hoặc cực tiểu (Infinity)
    if np.isinf(data_train_merged[col]).any():
        # Tìm giá trị lớn nhất MÀ KHÔNG PHẢI VÔ CỰC của cột đó
        max_val = data_train_merged.loc[~np.isinf(data_train_merged[col]), col].max()
        min_val = data_train_merged.loc[~np.isinf(data_train_merged[col]), col].min()
        
        # Nếu vô tình toàn bộ dữ liệu cột đó đều là Inf -> Điền 0, nếu không thì cắt trần bằng max/min thực tế
        max_val = max_val if pd.notna(max_val) else 0.0
        min_val = min_val if pd.notna(min_val) else 0.0
        
        # Thay thế Inf bằng max và -Inf bằng min
        data_train_merged.loc[:, col] = data_train_merged[col].replace(np.inf, max_val).replace(-np.inf, min_val)

# 2. XÓA NHÃN ĐỂ SCALE
if 'label' in numerical_cols:
    numerical_cols.remove('label')

y_train = data_train_merged['label']

# 3. FIT VÀ TRANSFORM ROBUST SCALER CHO TẬP TRAIN
print("\nĐang Scale tập TRAIN bằng StandardScaler...")
scaler = QuantileTransformer(output_distribution='normal', random_state=42)
data_train_base_merged_robust_scaler = data_train_merged.copy()
data_train_base_merged_robust_scaler[numerical_cols] = scaler.fit_transform(data_train_base_merged_robust_scaler[numerical_cols])

data_train_base_merged_robust_scaler['label'] = y_train

# 4. LƯU TẬP TRAIN
os.makedirs("merged-scaled-data", exist_ok=True)
train_save_path = os.path.join("merged-scaled-data", "data_train_merged_robust.parquet")
data_train_base_merged_robust_scaler.to_parquet(train_save_path, index=False, engine='pyarrow')
print(f"Đã lưu tập Train (Scaled) vào: {train_save_path}")

del data_train_merged, data_train_base_merged_robust_scaler
gc.collect()

# 5. ĐỌC VÀ GỘP TẬP TEST
print("\nBắt đầu đọc và gộp (merge) tập TEST...")
test_files = glob.glob(r'ip-files-test\*.parquet')
test_dfs = [pd.read_parquet(f) for f in test_files]
data_test_merged = pd.concat(test_dfs, ignore_index=True)

del test_dfs
gc.collect()
print(f"Tổng kích thước tập Test: {data_test_merged.shape}")

# XỬ LÝ LỖI INFINITY (Cho Test)
print("Đang dọn dẹp các giá trị Infinity (np.inf) trong tập Test...")
for col in numerical_cols:
    if np.isinf(data_test_merged[col]).any():
        # Xóa inf bằng max thực tế
        max_val = data_test_merged.loc[~np.isinf(data_test_merged[col]), col].max()
        min_val = data_test_merged.loc[~np.isinf(data_test_merged[col]), col].min()
        
        max_val = max_val if pd.notna(max_val) else 0.0
        min_val = min_val if pd.notna(min_val) else 0.0
        data_test_merged.loc[:, col] = data_test_merged[col].replace(np.inf, max_val).replace(-np.inf, min_val)

y_test = data_test_merged['label']

# 6. TRANSFORM TẬP TEST
print("Đang Scale tập TEST...")
data_test_merged_robust_scaler = data_test_merged.copy()
data_test_merged_robust_scaler[numerical_cols] = scaler.transform(data_test_merged_robust_scaler[numerical_cols])
data_test_merged_robust_scaler['label'] = y_test

# 7. LƯU TẬP TEST
test_save_path = os.path.join("merged-scaled-data", "data_test_merged_robust.parquet")
data_test_merged_robust_scaler.to_parquet(test_save_path, index=False, engine='pyarrow')
print(f"Đã lưu tập Test (Scaled) vào: {test_save_path}")

# Đọc và gộp tập valid_base
print("\nBắt đầu đọc và gộp (merge) tập VALID_BASE...")
valid_base_files = glob.glob(r'ip-files-valid-base\*.parquet')
valid_base_dfs = [pd.read_parquet(f) for f in valid_base_files]
data_valid_base_merged = pd.concat(valid_base_dfs, ignore_index=True)
del valid_base_dfs

# xử lý infinity cho valid_base
print("Đang dọn dẹp các giá trị Infinity (np.inf) trong tập Valid_base...")
for col in numerical_cols:
    if np.isinf(data_valid_base_merged[col]).any():
        max_val = data_valid_base_merged.loc[~np.isinf(data_valid_base_merged[col]), col].max()
        min_val = data_valid_base_merged.loc[~np.isinf(data_valid_base_merged[col]), col].min()
        
        max_val = max_val if pd.notna(max_val) else 0.0
        min_val = min_val if pd.notna(min_val) else 0.0
        data_valid_base_merged.loc[:, col] = data_valid_base_merged[col].replace(np.inf, max_val).replace(-np.inf, min_val)
y_valid = data_valid_base_merged['label']
data_valid_base_merged_robust_scaler = data_valid_base_merged.copy()
data_valid_base_merged_robust_scaler[numerical_cols] = scaler.transform(data_valid_base_merged_robust_scaler[numerical_cols])
data_valid_base_merged_robust_scaler['label'] = y_valid
valid_save_path = os.path.join("merged-scaled-data", "data_valid_base_merged_robust.parquet")
data_valid_base_merged_robust_scaler.to_parquet(valid_save_path, index=False, engine='pyarrow')
print(f"Đã lưu tập Valid_base (Scaled) vào: {valid_save_path}")

# Đọc và gộp tập train_meta
print("\nBắt đầu đọc và gộp (merge) tập TRAIN_META...")
train_meta_files = glob.glob(r'ip-files-train-meta\*.parquet')
train_meta_dfs = [pd.read_parquet(f) for f in train_meta_files]
data_train_meta_merged = pd.concat(train_meta_dfs, ignore_index=True)
del train_meta_dfs
print(f"Tổng kích thước tập Train_meta: {data_train_meta_merged.shape}")
print("Đang dọn dẹp các giá trị Infinity (np.inf) trong tập Train_meta...")
for col in numerical_cols:
    if np.isinf(data_train_meta_merged[col]).any():
        max_val = data_train_meta_merged.loc[~np.isinf(data_train_meta_merged[col]), col].max()
        min_val = data_train_meta_merged.loc[~np.isinf(data_train_meta_merged[col]), col].min()
        
        max_val = max_val if pd.notna(max_val) else 0.0
        min_val = min_val if pd.notna(min_val) else 0.0
        data_train_meta_merged.loc[:, col] = data_train_meta_merged[col].replace(np.inf, max_val).replace(-np.inf, min_val)
y_train_meta = data_train_meta_merged['label']
data_train_meta_merged_robust_scaler = data_train_meta_merged.copy()
data_train_meta_merged_robust_scaler[numerical_cols] = scaler.transform(data_train_meta_merged_robust_scaler[numerical_cols])
data_train_meta_merged_robust_scaler['label'] = y_train_meta
train_meta_save_path = os.path.join("merged-scaled-data", "data_train_meta_merged_robust.parquet")
data_train_meta_merged_robust_scaler.to_parquet(train_meta_save_path, index=False, engine='pyarrow')
print(f"Đã lưu tập Train_meta (Scaled) vào: {train_meta_save_path}")


gc.collect()
print("\nHoàn tất !!!")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np
from sklearn.model_selection import train_test_split

# --- CẤU HÌNH ---
TIME_STEPS = 10
BATCH_SIZE = 256
VAL_RATIO = 0.2
NUM_CLASSES = len(le.classes_)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Đang sử dụng thiết bị: {DEVICE}")


# 1. Định nghĩa lớp Dataset cắt cửa sổ trượt siêu nhẹ
class SlidingWindowDataset(Dataset):
    def __init__(self, df, time_steps):
        # Tách features và label, chuyển thẳng thành PyTorch Tensors
        self.X = torch.tensor(df.drop(columns=['label']).values, dtype=torch.float32)
        self.y = torch.tensor(df['label'].values, dtype=torch.long)
        self.time_steps = time_steps
        
    def __len__(self):
        # Tổng số cửa sổ có thể trượt được
        return len(self.X) - self.time_steps + 1
        
    def __getitem__(self, idx):
        # Lấy 1 cửa sổ (từ idx đến idx + time_steps)
        window_X = self.X[idx : idx + self.time_steps]
        # Nhãn là dòng cuối cùng của cửa sổ
        label_y = self.y[idx + self.time_steps - 1]
        
        return window_X, label_y

print("Đang khởi tạo Dataset và DataLoader...")

# 2. Tạo đối tượng Dataset
train_dataset = SlidingWindowDataset(train_real_df, TIME_STEPS)
val_loader = SlidingWindowDataset(val_df, TIME_STEPS)
test_dataset = SlidingWindowDataset(test_df, TIME_STEPS)

# Xóa Dataframe Pandas để giải phóng RAM
import gc
del train_df, test_df
gc.collect()

# 3. Tạo DataLoader
# Tham số shuffle=True ở train_loader sẽ tự động xáo trộn các cửa sổ mỗi epoch
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)
val_loader = DataLoader(val_loader, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

NUM_FEATURES = train_dataset.X.shape[1]
print(f"Số lượng features: {NUM_FEATURES}")
```

5. Xay dựng mô hình CNN-BiLSTM bằng pytorch

```python
class CNN_BiLSTM(nn.Module):
    def __init__(self, num_features, num_classes, time_steps, hidden_size=128):
        super(CNN_BiLSTM, self).__init__()
        
        # --- Tầng 1D-CNN ---
        # in_channels chính là số lượng features (cột)
        self.conv1 = nn.Conv1d(in_channels=num_features, out_channels=64, kernel_size=3)
        self.bn1 = nn.BatchNorm1d(64)
        self.relu = nn.ReLU()
        self.pool1 = nn.MaxPool1d(kernel_size=2)
        
        # --- Tầng BiLSTM ---
        # Đầu vào của LSTM lúc này là 64 (do out_channels của CNN là 64)
        self.bilstm = nn.LSTM(input_size=64, hidden_size=hidden_size, 
                              batch_first=True, bidirectional=True)
        self.dropout = nn.Dropout(0.3)
        
        # --- Tầng Phân loại (Dense) ---
        # hidden_size * 2 vì là Bidirectional (2 chiều xuôi ngược)
        self.fc1 = nn.Linear(hidden_size * 2, 64)
        self.fc2 = nn.Linear(64, num_classes)
        
    def forward(self, x):
        # x đang có shape: (Batch, TimeSteps, Features)
        
        # 1. Đảo trục cho Conv1d -> (Batch, Features, TimeSteps)
        x = x.permute(0, 2, 1) 
        
        # Đi qua CNN
        x = self.conv1(x)
        x = self.bn1(x)
        x = self.relu(x)
        x = self.pool1(x)
        
        # 2. Đảo trục lại cho LSTM -> (Batch, NewTimeSteps, Channels)
        x = x.permute(0, 2, 1)
        
        # Đi qua BiLSTM
        out, (hn, cn) = self.bilstm(x)
        
        # 3. Lấy output của TimeStep cuối cùng để dự đoán
        # out shape: (Batch, SeqLen, Hidden*2) -> Lấy [:, -1, :]
        out = out[:, -1, :]
        
        # Đi qua Dense
        out = self.dropout(out)
        out = self.fc1(out)
        out = self.relu(out)
        out = self.dropout(out)
        out = self.fc2(out)
        return out

# Khởi tạo mô hình và đẩy lên GPU
model = CNN_BiLSTM(num_features=NUM_FEATURES, num_classes=NUM_CLASSES, time_steps=TIME_STEPS).to(DEVICE)
print(model)
```

6. Training mô hình (class_weights + sliding window)

```python
# in ra phân bổ label trong train_dataset
train_labels = train_dataset.y.numpy()
unique, counts = np.unique(train_labels, return_counts=True)
class_count_dict = dict(zip(unique, counts))

num_classes = len(class_count_dict)
counts = np.array([class_count_dict[i] for i in range(num_classes)])
total_samples = np.sum(counts)
class_weights = total_samples / (num_classes * counts)
smoothed_weights = np.sqrt(class_weights)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
class_weights_tensor = torch.tensor(smoothed_weights, dtype=torch.float32).to(device)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0):
        super(FocalLoss, self).__init__()
        self.alpha = alpha # Có thể truyền smoothed_weights vào đây
        self.gamma = gamma

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, weight=self.alpha, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = ((1 - pt) ** self.gamma) * ce_loss
        return focal_loss.mean()

# Khởi tạo (Bạn có thể kết hợp với smoothed_weights ở Cách 1)
criterion = FocalLoss(alpha=class_weights_tensor, gamma=2.0)
```

```python
from tqdm import tqdm
from sklearn.metrics import classification_report

# Khai báo Hàm Loss và Optimizer
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Learning Rate Scheduler (Giảm LR nếu loss không cải thiện)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2, verbose=True)

EPOCHS = 30
best_val_loss = float('inf')
patience_counter = 0
EARLY_STOP_PATIENCE = 5

for epoch in range(EPOCHS):
    # --- TRAINING ---
    model.train()
    train_loss = 0.0
    correct_train = 0
    total_train = 0
    
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS} [Train]", leave=False)
    for inputs, labels in loop:
        inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
        
        optimizer.zero_grad() # Xóa gradient cũ
        
        outputs = model(inputs) # Forward
        loss = criterion(outputs, labels) # Tính Loss
        
        loss.backward() # Lan truyền ngược
        optimizer.step() # Cập nhật trọng số
        
        train_loss += loss.item() * inputs.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total_train += labels.size(0)
        correct_train += (predicted == labels).sum().item()
        
        loop.set_postfix(loss=loss.item())

    avg_train_loss = train_loss / total_train
    train_acc = correct_train / total_train
    
    # --- VALIDATION / TESTING ---
    # Trong thực tế bạn nên có một tập Validation riêng, 
    # ở đây dùng tạm test_loader để in tiến độ (nên tách riêng lúc làm báo cáo)
    model.eval()
    test_loss = 0.0
    correct_test = 0
    total_test = 0
    
    all_preds = []
    all_targets = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(val_loader, desc=f"Epoch {epoch+1}/{EPOCHS} [Test]", leave=False):
            inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
            
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            
            test_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs.data, 1)
            
            total_test += labels.size(0)
            correct_test += (predicted == labels).sum().item()
            
            all_preds.extend(predicted.cpu().numpy())
            all_targets.extend(labels.cpu().numpy())

    avg_val_loss = test_loss / total_test
    val_acc = correct_test / total_test
    
    print(f"Epoch [{epoch+1}/{EPOCHS}] - Train Loss: {avg_train_loss:.4f}, Train Acc: {train_acc:.4f} | Test Loss: {avg_val_loss:.4f}, Test Acc: {val_acc:.4f}")
    
    # Kích hoạt Learning Rate Scheduler
    scheduler.step(avg_val_loss)
    
    # Early Stopping
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        patience_counter = 0
        # Lưu lại tệp trọng số tốt nhất
        torch.save(model.state_dict(), "best_cnn_bilstm.pth")
    else:
        patience_counter += 1
        if patience_counter >= EARLY_STOP_PATIENCE:
            print(f"\n[Early Stopping] Model không cải thiện sau {EARLY_STOP_PATIENCE} epochs. Dừng huấn luyện.")
            break

# --- IN BÁO CÁO CUỐI CÙNG ---
print("\n--- BÁO CÁO PHÂN LOẠI TRÊN TẬP TEST ---")
model.load_state_dict(torch.load("best_cnn_bilstm.pth"))
model.eval()

all_preds = []
all_targets = []

with torch.no_grad():
    for inputs, labels in tqdm(test_loader, desc="[Final Test]", leave=False):
        inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        all_preds.extend(predicted.cpu().numpy())
        all_targets.extend(labels.cpu().numpy())

print(classification_report(all_targets, all_preds, target_names=le.classes_))
```

7. Undersampling + Sliding Window

```python
import torch
from torch.utils.data import Dataset
import numpy as np
TIME_STEPS = 100
BATCH_SIZE = 256
VAL_RATIO = 0.2
NUM_CLASSES = len(le.classes_)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Đang sử dụng thiết bị: {DEVICE}")


# 1. Định nghĩa lớp Dataset cắt cửa sổ trượt siêu nhẹ
class SlidingWindowDataset(Dataset):
    def __init__(self, df, time_steps, max_samples_per_class = 20000, stride = 20):
        # Tách features và label, chuyển thẳng thành PyTorch Tensors
        self.X = torch.tensor(df.drop(columns=['label']).values, dtype=torch.float32)
        self.y = torch.tensor(df['label'].values, dtype=torch.long)
        self.time_steps = time_steps

        num_possible_windows = len(self.X) - self.time_steps + 1
        all_possible_indices = np.arange(0, num_possible_windows, stride)

        window_labels = self.y[all_possible_indices + self.time_steps -1].numpy()
        self.valid_indices = []

        classes, counts = np.unique(window_labels, return_counts=True)
        
        for c, count in zip(classes, counts):
            c_indices = all_possible_indices[window_labels == c]
            
            if count > max_samples_per_class:
                c_indices = np.random.choice(c_indices, max_samples_per_class, replace=False)
                print(f"Class {c}: Giảm từ {count} xuống {max_samples_per_class} cửa sổ.")
            else:
                print(f"Class {c}: Giữ nguyên {count} cửa sổ.")
                
            self.valid_indices.extend(c_indices)
            
        np.random.shuffle(self.valid_indices)
        
    def __len__(self):
        # Tổng số cửa sổ có thể trượt được
        return len(self.valid_indices)
        
    def __getitem__(self, idx):
        actual_idx = self.valid_indices[idx]
        window_X = self.X[actual_idx : actual_idx + self.time_steps]
        label_y = self.y[actual_idx + self.time_steps - 1]
        return window_X, label_y
```

```python
import torch
from torch.utils.data import Dataset
import numpy as np

class UndersampledSlidingWindowDataset(Dataset):
    def __init__(self, df, time_steps, max_samples_per_class=20000):
        """
        max_samples_per_class: Số lượng cửa sổ tối đa được giữ lại cho mỗi class.
        Class nào có số mẫu > ngưỡng này sẽ bị Random Undersample xuống.
        Class nào < ngưỡng này sẽ được giữ nguyên 100%.
        """
        self.X = torch.tensor(df.drop(columns=['label']).values, dtype=torch.float32)
        self.y = torch.tensor(df['label'].values, dtype=torch.long)
        self.time_steps = time_steps
        
        print("Đang tính toán các cửa sổ hợp lệ và Undersampling...")
        
        # 1. Lấy nhãn của tất cả các cửa sổ có thể tạo ra
        # Nhãn của cửa sổ trượt nằm ở vị trí (thời_điểm_hiện_tại + time_steps - 1)
        num_possible_windows = len(self.X) - self.time_steps + 1
        window_labels = self.y[self.time_steps - 1 : ].numpy()
        all_indices = np.arange(num_possible_windows)
        
        self.valid_indices = []
        
        # 2. Lặp qua từng class để chọn mẫu
        classes, counts = np.unique(window_labels, return_counts=True)
        
        for c, count in zip(classes, counts):
            # Lấy tất cả index của các cửa sổ thuộc class c
            c_indices = all_indices[window_labels == c]
            
            if count > max_samples_per_class:
                # Nếu quá nhiều -> Random Undersampling (chọn ngẫu nhiên không hoàn lại)
                c_indices = np.random.choice(c_indices, max_samples_per_class, replace=False)
                print(f"Class {c}: Giảm từ {count} xuống {max_samples_per_class} cửa sổ.")
            else:
                # Nếu ít -> Giữ nguyên
                print(f"Class {c}: Giữ nguyên {count} cửa sổ.")
                
            self.valid_indices.extend(c_indices)
            
        # 3. Trộn đều danh sách index để các batch đa dạng hơn
        np.random.shuffle(self.valid_indices)
        print(f"Tổng số cửa sổ sau khi Undersampling: {len(self.valid_indices)}")

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        # Lấy index thực sự từ danh sách đã được lọc
        actual_idx = self.valid_indices[idx]
        
        # Cắt đúng cửa sổ liên tục không bị đứt gãy
        window_X = self.X[actual_idx : actual_idx + self.time_steps]
        label_y = self.y[actual_idx + self.time_steps - 1]
        
        return window_X, label_y
```

```python
import pandas as pd

# --- CẤU HÌNH ---
VAL_RATIO = 0.2

def stratified_time_series_split(df, target_col, val_ratio=0.2):
    train_indices = []
    val_indices = []
    
    # Lấy danh sách các class duy nhất
    classes = df[target_col].unique()
    
    for c in classes:
        # Lấy tất cả index của class này
        class_indices = df[df[target_col] == c].index.tolist()
        
        # Tính toán điểm cắt (không shuffle, giữ nguyên thứ tự thời gian của class đó)
        split_idx = int(len(class_indices) * (1 - val_ratio))
        
        train_indices.extend(class_indices[:split_idx])
        val_indices.extend(class_indices[split_idx:])
    
    # Sắp xếp lại index để đảm bảo tính tuần tự của thời gian trên toàn bộ dataset (tùy chọn)
    train_indices.sort()
    val_indices.sort()
    
    # Tạo DataFrame mới từ các index đã chia
    train_real_df = df.loc[train_indices].reset_index(drop=True)
    val_df = df.loc[val_indices].reset_index(drop=True)
    
    return train_real_df, val_df

# Thực hiện chia
print("Đang chia tập dữ liệu theo Stratified Time-Series...")
train_real_df, val_df = stratified_time_series_split(train_df, 'label', VAL_RATIO)

# Kiểm tra tỷ lệ để đảm bảo stratify hoạt động
print(f"Tỷ lệ train: {len(train_real_df)} dòng")
print(f"Tỷ lệ val: {len(val_df)} dòng")
print("\nPhân bổ class trong tập Val:\n", val_df['label'].value_counts(normalize=True))

```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0):
        super(FocalLoss, self).__init__()
        self.alpha = alpha # Có thể truyền smoothed_weights vào đây
        self.gamma = gamma

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, weight=self.alpha, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = ((1 - pt) ** self.gamma) * ce_loss
        return focal_loss.mean()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
NUM_FEATURES = train_dataset.X.shape[1]
# ==========================================
# 1. Định nghĩa Module Attention Đơn giản
# ==========================================
class Attention(nn.Module):
    def __init__(self, hidden_dim):
        super(Attention, self).__init__()
        # Lớp Linear để tính điểm số (score) cho từng time-step
        self.attention = nn.Linear(hidden_dim, 1)

    def forward(self, lstm_outputs):
        # lstm_outputs shape: (Batch, SeqLen, Hidden*2)
        
        # 1. Tính điểm chú ý cho mỗi timestep -> Shape: (Batch, SeqLen, 1)
        scores = self.attention(lstm_outputs)
        
        # 2. Áp dụng Softmax dọc theo trục SeqLen để lấy trọng số (tổng các timestep = 1)
        weights = torch.softmax(scores, dim=1)
        
        # 3. Nhân trọng số với output của LSTM và tính tổng (Weighted Sum)
        # (Batch, SeqLen, 1) * (Batch, SeqLen, Hidden*2) -> (Batch, Hidden*2)
        context_vector = torch.sum(weights * lstm_outputs, dim=1)
        
        return context_vector, weights

# ==========================================
# 2. Định nghĩa Kiến trúc Chính
# ==========================================
class CNN_BiLSTM_Attention(nn.Module):
    def __init__(self, num_features, num_classes, time_steps, hidden_size=128):
        super(CNN_BiLSTM_Attention, self).__init__()
        
        # --- Tầng 1D-CNN (Nâng cấp Sức chứa) ---
        # Block 1: Tăng từ num_features lên 128 (Thêm padding=1 để không mất chiều dài chuỗi quá nhanh)
        self.conv1 = nn.Conv1d(in_channels=num_features, out_channels=64, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm1d(64)
        self.dropout_cnn = nn.Dropout1d(0.2)

        # Block 2 (Mới): Lọc sâu thêm 1 lớp nữa, tăng kênh lên 256
        self.conv2 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm1d(128)
        
        self.relu = nn.ReLU()
        
        # Pooling gộp 2 bước lại làm 1 để giảm tải cho LSTM
        self.pool = nn.MaxPool1d(kernel_size=2) 
        
        # --- Tầng BiLSTM ---
        # Đầu vào của LSTM giờ là 256 (do out_channels của lớp Conv2)
        self.bilstm = nn.LSTM(input_size=128, hidden_size=hidden_size, 
                              batch_first=True, bidirectional=True)
        
        # --- Tầng Attention (MỚI) ---
        # Kích thước đầu vào của Attention = hidden_size * 2 (vì BiLSTM ghép 2 chiều ngược xuôi)
        self.attention = Attention(hidden_size * 2)
        
        self.dropout = nn.Dropout(0.5)
        
        # --- Tầng Phân loại (Dense) ---
        self.fc1 = nn.Linear(hidden_size * 2, 64)
        self.fc2 = nn.Linear(64, num_classes)
        
    def forward(self, x):
        # Đầu vào x: (Batch, TimeSteps, Features)
        
        # 1. Đảo trục cho CNN -> (Batch, Features, TimeSteps)
        x = x.permute(0, 2, 1) 
        
        # Đi qua CNN Block 1
        x = self.conv1(x)
        x = self.bn1(x)
        x = self.relu(x)
        x = self.dropout_cnn(x)

        # Đi qua CNN Block 2
        x = self.conv2(x)
        x = self.bn2(x)
        x = self.relu(x)
        
        # Pooling giảm chiều dài chuỗi
        x = self.pool(x)
        
        # 2. Đảo trục lại cho LSTM -> (Batch, NewTimeSteps, 256 Channels)
        x = x.permute(0, 2, 1)
        
        # Đi qua BiLSTM
        # out shape: (Batch, SeqLen, Hidden*2)
        out, (hn, cn) = self.bilstm(x)
        
        # 3. CƠ CHẾ ATTENTION (Thay vì lấy out[:, -1, :])
        # context_vector đã tổng hợp thông tin quan trọng nhất của toàn bộ cửa sổ
        context_vector, attn_weights = self.attention(out)
        
        # 4. Đi qua Dense
        out = self.dropout(context_vector)
        out = self.fc1(out)
        out = self.relu(out)
        out = self.dropout(out)
        out = self.fc2(out)
        
        return out

# Khởi tạo mô hình mới và đẩy lên thiết bị
model = CNN_BiLSTM_Attention(num_features=NUM_FEATURES, num_classes=NUM_CLASSES, time_steps=TIME_STEPS).to(DEVICE)
print(model)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
NUM_FEATURES = train_dataset.X.shape[1]
# ==========================================
# 1. Cơ chế Attention
# ==========================================
class Attention(nn.Module):
    def __init__(self, hidden_dim):
        super(Attention, self).__init__()
        self.attention = nn.Linear(hidden_dim, 1)

    def forward(self, lstm_outputs):
        # Tính điểm chú ý
        scores = self.attention(lstm_outputs)
        weights = torch.softmax(scores, dim=1)
        # Nhân trọng số và tính tổng
        context_vector = torch.sum(weights * lstm_outputs, dim=1)
        return context_vector, weights

# ==========================================
# 2. Kiến trúc chính bám sát Bài Báo
# ==========================================
class Paper_CNN_BiLSTM_Attention(nn.Module):
    def __init__(self, num_features, num_classes):
        super(Paper_CNN_BiLSTM_Attention, self).__init__()
        
        # --- Tầng CNN Block 1 ---
        # Bài báo: 128 filters, kernel size 5, ReLU [cite: 119] -> Batch Norm, MaxPool(2), Dropout(0.3) [cite: 120, 167]
        # in_channels = 1 (Vì ta coi 347 features như 1 chuỗi dài 347 đơn vị)
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=128, kernel_size=5, padding=2)
        self.bn1 = nn.BatchNorm1d(128)
        self.pool1 = nn.MaxPool1d(kernel_size=2)
        self.drop1 = nn.Dropout1d(0.3)

        # --- Tầng CNN Block 2 ---
        # Bài báo: 256 filters, kernel size 5, ReLU -> Batch Norm, MaxPool(2), Dropout(0.3) [cite: 121, 168, 169]
        self.conv2 = nn.Conv1d(in_channels=128, out_channels=256, kernel_size=5, padding=2)
        self.bn2 = nn.BatchNorm1d(256)
        self.pool2 = nn.MaxPool1d(kernel_size=2)
        self.drop2 = nn.Dropout1d(0.3)

        # --- Tầng CNN Block 3 ---
        # Bài báo: 128 filters, kernel size 3, ReLU -> Batch Norm, MaxPool(2), Dropout(0.3) [cite: 121, 169, 170]
        self.conv3 = nn.Conv1d(in_channels=256, out_channels=128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm1d(128)
        self.pool3 = nn.MaxPool1d(kernel_size=2)
        self.drop3 = nn.Dropout1d(0.3)

        # --- Tầng BiLSTM ---
        # Bài báo: Bi-LSTM 128 units [cite: 122], theo sau là Dropout(0.3) [cite: 171]
        # Đầu vào của LSTM là 128 channels từ conv3
        self.bilstm = nn.LSTM(input_size=128, hidden_size=128, batch_first=True, bidirectional=True)
        self.drop_lstm = nn.Dropout(0.3)

        # --- Tầng Attention ---
        # Bài báo: Attention mechanism sau LSTM [cite: 122, 171]
        self.attention = Attention(128 * 2) # *2 vì là Bidirectional

        # --- Các Tầng Dense (Fully Connected) ---
        # Bài báo: Dense 256 + ReLU -> Dropout(0.4) [cite: 123, 171]
        self.fc1 = nn.Linear(128 * 2, 256)
        self.drop_fc1 = nn.Dropout(0.4)

        # Bài báo: Dense 128 + ReLU -> Dropout(0.4) [cite: 123, 172]
        self.fc2 = nn.Linear(256, 128)
        self.drop_fc2 = nn.Dropout(0.4)

        # Output layer: 10 class Softmax [cite: 124, 173] (Ta dùng num_classes của bạn = 11)
        self.fc3 = nn.Linear(128, num_classes)

    def forward(self, x):
        # Đầu vào x từ StandardFlowDataset đang có shape: (Batch, Features)
        
        # MẸO QUAN TRỌNG: Thêm chiều Channel = 1 để đưa vào Conv1d
        # Shape trở thành: (Batch, 1, Features)
        x = x.unsqueeze(1) 

        # --- Đi qua 3 Block CNN ---
        x = self.conv1(x)
        x = self.bn1(x)
        x = F.relu(x)
        x = self.pool1(x)
        x = self.drop1(x)

        x = self.conv2(x)
        x = self.bn2(x)
        x = F.relu(x)
        x = self.pool2(x)
        x = self.drop2(x)

        x = self.conv3(x)
        x = self.bn3(x)
        x = F.relu(x)
        x = self.pool3(x)
        x = self.drop3(x)

        # --- Chuẩn bị cho BiLSTM ---
        # Shape hiện tại: (Batch, Channels=128, NewLength)
        # BiLSTM cần shape: (Batch, SequenceLength, Features/Channels)
        x = x.permute(0, 2, 1)

        # --- Đi qua BiLSTM ---
        out, _ = self.bilstm(x)
        out = self.drop_lstm(out)

        # --- Đi qua Attention ---
        context_vector, _ = self.attention(out)

        # --- Đi qua Dense Layers ---
        out = self.fc1(context_vector)
        out = F.relu(out)
        out = self.drop_fc1(out)

        out = self.fc2(out)
        out = F.relu(out)
        out = self.drop_fc2(out)

        out = self.fc3(out)
        
        # Lưu ý: Không dùng F.softmax ở đây vì hàm nn.CrossEntropyLoss() đã tự động tính Softmax bên trong nó.
        return out

# Khởi tạo mô hình và đẩy lên GPU
model = Paper_CNN_BiLSTM_Attention(num_features=NUM_FEATURES, num_classes=NUM_CLASSES).to(DEVICE)
print(model)
```

```python
from torch.utils.data import Dataset
class StandardFlowDataset(Dataset):
    def __init__(self, df, max_samples_per_class=None):
        self.X = []
        self.y = []
        
        # Nhóm theo từng class để Undersample
        for class_name, group in df.groupby('label'):
            if max_samples_per_class and len(group) > max_samples_per_class:
                group = group.sample(n=max_samples_per_class, random_state=42)
            
            self.X.append(group.drop(columns=['label']).values)
            self.y.append(group['label'].values)
            
        self.X = np.vstack(self.X)
        self.y = np.concatenate(self.y)
        
        # Xáo trộn dữ liệu
        idx = np.random.permutation(len(self.X))
        self.X = torch.tensor(self.X[idx], dtype=torch.float32)
        self.y = torch.tensor(self.y[idx], dtype=torch.long)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# Khởi tạo lại
MAX_SAMPLES = 20000 
train_dataset = StandardFlowDataset(train_real_df, max_samples_per_class=MAX_SAMPLES)
val_dataset = StandardFlowDataset(val_df) # Không truyền max_samples để giữ nguyên
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class Attention(nn.Module):
    def __init__(self, hidden_dim):
        super(Attention, self).__init__()
        self.attention = nn.Linear(hidden_dim, 1)

    def forward(self, lstm_outputs):
        scores = self.attention(lstm_outputs)
        weights = torch.softmax(scores, dim=1)
        context_vector = torch.sum(weights * lstm_outputs, dim=1)
        return context_vector, weights

class Hybrid_DNN_BiLSTM_Attention(nn.Module):
    def __init__(self, num_features, num_classes):
        super(Hybrid_DNN_BiLSTM_Attention, self).__init__()
        
        # ==========================================
        # 1. KHỐI DNN (Thay thế 1D-CNN)
        # ==========================================
        self.dnn_extractor = nn.Sequential(
            nn.Linear(num_features, 512),
            nn.BatchNorm1d(512),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.1),
            
            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.1),
            
            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.LeakyReLU(0.1)
        )

        # ==========================================
        # 2. KHỐI BiLSTM
        # ==========================================
        # Nhận 128 features từ khối DNN
        self.bilstm = nn.LSTM(input_size=128, hidden_size=128, batch_first=True, bidirectional=True)
        self.drop_lstm = nn.Dropout(0.1)

        # ==========================================
        # 3. KHỐI ATTENTION & OUTPUT
        # ==========================================
        self.attention = Attention(128 * 2) # *2 vì BiLSTM

        self.fc1 = nn.Linear(128 * 2, 128)
        self.drop_fc1 = nn.Dropout(0.1)
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        # x shape ban đầu: (Batch, 347)
        
        # Bước 1: Trích xuất đặc trưng bằng DNN
        features = self.dnn_extractor(x) # shape: (Batch, 128)
        
        # Bước 2: Thêm chiều Sequence Length = 1 để "lừa" LSTM
        # Shape trở thành: (Batch, 1, 128)
        lstm_input = features.unsqueeze(1) 
        
        # Bước 3: Đưa vào BiLSTM
        lstm_out, _ = self.bilstm(lstm_input) # lstm_out shape: (Batch, 1, 256)
        lstm_out = self.drop_lstm(lstm_out)
        
        # Bước 4: Đưa qua Attention
        context_vector, _ = self.attention(lstm_out) # context_vector shape: (Batch, 256)
        
        # Bước 5: Phân loại cuối cùng
        out = self.fc1(context_vector)
        out = F.leaky_relu(out, 0.1)
        out = self.drop_fc1(out)
        out = self.fc2(out)
        
        return out

# Khởi tạo mô hình
model = Hybrid_DNN_BiLSTM_Attention(num_features=NUM_FEATURES, num_classes=NUM_CLASSES).to(DEVICE)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# ==========================================
# BƯỚC 1: LỌC NHIỄU BẰNG RANDOM FOREST
# ==========================================
print("🌲 Đang dùng Random Forest chọn 100 cột tinh túy nhất...")
X_rf = train_dataset.X.numpy()
y_rf = train_dataset.y.numpy()

rf = RandomForestClassifier(n_estimators=50, random_state=42, n_jobs=-1)
rf.fit(X_rf, y_rf)

# Lấy 100 cột quan trọng nhất
top_100_idx = np.argsort(rf.feature_importances_)[-100:]

# Cập nhật Dataset: Gọt bỏ 247 cột rác
train_dataset.X = train_dataset.X[:, top_100_idx]
val_dataset.X = val_dataset.X[:, top_100_idx]
NUM_FEATURES = 100

print(f"✅ Đã gọt dữ liệu! Kích thước đầu vào hiện tại: {train_dataset.X.shape}")

# Tăng Batch Size lên 256 để Gradient cực kỳ mượt mà
BATCH_SIZE = 256
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

# ==========================================
# BƯỚC 2: KHỞI TẠO MÔ HÌNH HYBRID DNN
# ==========================================
class Attention(nn.Module):
    def __init__(self, hidden_dim):
        super(Attention, self).__init__()
        self.attention = nn.Linear(hidden_dim, 1)

    def forward(self, lstm_outputs):
        scores = self.attention(lstm_outputs)
        weights = torch.softmax(scores, dim=1)
        context_vector = torch.sum(weights * lstm_outputs, dim=1)
        return context_vector, weights

class Slim_Hybrid_DNN_BiLSTM_Attention(nn.Module):
    def __init__(self, num_features, num_classes):
        super(Slim_Hybrid_DNN_BiLSTM_Attention, self).__init__()
        
        # Khối DNN đã được "ép cân" và tăng cường phòng thủ
        self.dnn = nn.Sequential(
            nn.Linear(num_features, 256), # Giảm từ 512 xuống 256
            nn.BatchNorm1d(256),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.3), # TĂNG Dropout lên 0.3 để chống Overfitting
            
            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.3)  # TĂNG Dropout
        )

        self.bilstm = nn.LSTM(128, 128, batch_first=True, bidirectional=True)
        self.attention = Attention(256)
        self.fc = nn.Linear(256, num_classes)

    def forward(self, x):
        features = self.dnn(x)
        lstm_out, _ = self.bilstm(features.unsqueeze(1)) 
        context_vector, _ = self.attention(lstm_out)
        return self.fc(context_vector)

model = Slim_Hybrid_DNN_BiLSTM_Attention(num_features=NUM_FEATURES, num_classes=NUM_CLASSES).to(DEVICE)

# ==========================================
# BƯỚC 3: CẤU HÌNH HUẤN LUYỆN CHUẨN MỰC
# ==========================================
# KHÔNG dùng Label Smoothing nữa (DNN đủ sức phân biệt)
criterion = nn.CrossEntropyLoss()

# Dùng AdamW với tốc độ chậm, ổn định
optimizer = torch.optim.AdamW(model.parameters(), lr=0.0002, weight_decay=1e-3)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)
EPOCHS = 50
EARLY_STOP_PATIENCE = 10 # Cho phép mô hình kiên nhẫn hơn một chút
best_val_f1 = 0.0
patience_counter = 0

print("🚀 BẮT ĐẦU HUẤN LUYỆN MÔ HÌNH DNN + BiLSTM + ATTENTION...")

# ==========================================
# 2. VÒNG LẶP HUẤN LUYỆN
# ==========================================
for epoch in range(EPOCHS):
    # --- TRAINING ---
    model.train()
    train_loss = 0.0
    
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS} [Train]", leave=False)
    for inputs, labels in loop:
        # Đảm bảo dữ liệu nằm trên GPU
        inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
        
        # Reset gradient
        optimizer.zero_grad()
        
        # Forward pass (Sinh dự đoán)
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        
        # Backward pass (Tính toán đạo hàm)
        loss.backward()
        
        # 🛡️ BẢO HIỂM CHỐNG NỔ LOSS (Gradient Clipping)
        # Giới hạn độ lớn của Gradient không vượt quá 1.0, ngăn chặn Dying ReLU và NaN
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        # Cập nhật trọng số
        optimizer.step()
        
        train_loss += loss.item() * inputs.size(0)
        loop.set_postfix(loss=loss.item())
        
    avg_train_loss = train_loss / len(train_loader.dataset)
    
    # --- VALIDATION ---
    model.eval()
    val_loss = 0.0
    all_preds = []
    all_targets = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(val_loader, desc=f"Epoch {epoch+1}/{EPOCHS} [Val]", leave=False):
            inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
            
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            
            val_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs.data, 1)
            
            all_preds.extend(predicted.cpu().numpy())
            all_targets.extend(labels.cpu().numpy())
            
    avg_val_loss = val_loss / len(val_loader.dataset)
    
    # Đánh giá bằng Macro-F1 (Công bằng cho mọi class)
    val_f1_macro = f1_score(all_targets, all_preds, average='macro')
    
    print(f"Epoch [{epoch+1}/{EPOCHS}] - Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f} | Val Macro-F1: {val_f1_macro:.4f}")
    
    # Điều chỉnh Learning Rate dựa trên F1
    scheduler.step()
    
    # --- EARLY STOPPING & BÁO CÁO ---
    if val_f1_macro > best_val_f1:
        best_val_f1 = val_f1_macro
        patience_counter = 0
        
        # Lưu lại tệp trọng số xuất sắc nhất
        torch.save(model.state_dict(), "best_paper_model.pth")
        print(f" -> 🎉 Đã lưu model tốt nhất mới với Val Macro-F1: {best_val_f1:.4f}\n")
    else:
        patience_counter += 1
        if patience_counter >= EARLY_STOP_PATIENCE:
            print(f"\n🛑 [Early Stopping] Mô hình không cải thiện F1 sau {EARLY_STOP_PATIENCE} epochs liên tiếp. Dừng huấn luyện.")
            break

print(f"🎯 Huấn luyện hoàn tất! Best Validation Macro-F1 đạt được: {best_val_f1:.4f}")
```

```python
from torch.utils.data import Dataset
import torch
import numpy as np
from torch.utils.data import DataLoader
class StandardFlowDataset(Dataset):
    def __init__(self, df, max_samples_per_class=None):
        self.X = []
        self.y = []
        
        # Nhóm theo từng class để Undersample
        for class_name, group in df.groupby('label'):
            if max_samples_per_class and len(group) > max_samples_per_class:
                group = group.sample(n=max_samples_per_class, random_state=42)
            
            self.X.append(group.drop(columns=['label']).values)
            self.y.append(group['label'].values)
            
        self.X = np.vstack(self.X)
        self.y = np.concatenate(self.y)
        
        # Xáo trộn dữ liệu
        idx = np.random.permutation(len(self.X))
        self.X = torch.tensor(self.X[idx], dtype=torch.float32)
        self.y = torch.tensor(self.y[idx], dtype=torch.long)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
NUM_FEATURES = train_dataset.X.shape[1]
NUM_CLASSES = len(train_dataset.y.unique())
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# ==========================================
# 1. Cơ chế Attention
# ==========================================
class Attention(nn.Module):
    def __init__(self, hidden_dim):
        super(Attention, self).__init__()
        self.attention = nn.Linear(hidden_dim, 1)

    def forward(self, lstm_outputs):
        # Tính điểm chú ý
        scores = self.attention(lstm_outputs)
        weights = torch.softmax(scores, dim=1)
        # Nhân trọng số và tính tổng
        context_vector = torch.sum(weights * lstm_outputs, dim=1)
        return context_vector, weights

# ==========================================
# 2. Kiến trúc chính bám sát Bài Báo
# ==========================================
class Paper_CNN_BiLSTM_Attention(nn.Module):
    def __init__(self, num_features, num_classes):
        super(Paper_CNN_BiLSTM_Attention, self).__init__()
        
        # --- Tầng CNN Block 1 ---
        # Bài báo: 128 filters, kernel size 5, ReLU [cite: 119] -> Batch Norm, MaxPool(2), Dropout(0.3) [cite: 120, 167]
        # in_channels = 1 (Vì ta coi 347 features như 1 chuỗi dài 347 đơn vị)
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=128, kernel_size=5, padding=2)
        self.bn1 = nn.BatchNorm1d(128)
        self.pool1 = nn.MaxPool1d(kernel_size=2)
        self.drop1 = nn.Dropout1d(0.3)

        # --- Tầng CNN Block 2 ---
        # Bài báo: 256 filters, kernel size 5, ReLU -> Batch Norm, MaxPool(2), Dropout(0.3) [cite: 121, 168, 169]
        self.conv2 = nn.Conv1d(in_channels=128, out_channels=256, kernel_size=5, padding=2)
        self.bn2 = nn.BatchNorm1d(256)
        self.pool2 = nn.MaxPool1d(kernel_size=2)
        self.drop2 = nn.Dropout1d(0.3)

        # --- Tầng CNN Block 3 ---
        # Bài báo: 128 filters, kernel size 3, ReLU -> Batch Norm, MaxPool(2), Dropout(0.3) [cite: 121, 169, 170]
        self.conv3 = nn.Conv1d(in_channels=256, out_channels=128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm1d(128)
        self.pool3 = nn.MaxPool1d(kernel_size=2)
        self.drop3 = nn.Dropout1d(0.3)

        # --- Tầng BiLSTM ---
        # Bài báo: Bi-LSTM 128 units [cite: 122], theo sau là Dropout(0.3) [cite: 171]
        # Đầu vào của LSTM là 128 channels từ conv3
        self.bilstm = nn.LSTM(input_size=128, hidden_size=128, batch_first=True, bidirectional=True)
        self.drop_lstm = nn.Dropout(0.3)

        # --- Tầng Attention ---
        # Bài báo: Attention mechanism sau LSTM [cite: 122, 171]
        self.attention = Attention(128 * 2) # *2 vì là Bidirectional

        # --- Các Tầng Dense (Fully Connected) ---
        # Bài báo: Dense 256 + ReLU -> Dropout(0.4) [cite: 123, 171]
        self.fc1 = nn.Linear(128 * 2, 256)
        self.drop_fc1 = nn.Dropout(0.4)

        # Bài báo: Dense 128 + ReLU -> Dropout(0.4) [cite: 123, 172]
        self.fc2 = nn.Linear(256, 128)
        self.drop_fc2 = nn.Dropout(0.4)

        # Output layer: 10 class Softmax [cite: 124, 173] (Ta dùng num_classes của bạn = 11)
        self.fc3 = nn.Linear(128, num_classes)

    def forward(self, x):
        # Đầu vào x từ StandardFlowDataset đang có shape: (Batch, Features)
        
        # MẸO QUAN TRỌNG: Thêm chiều Channel = 1 để đưa vào Conv1d
        # Shape trở thành: (Batch, 1, Features)
        x = x.unsqueeze(1) 

        # --- Đi qua 3 Block CNN ---
        x = self.conv1(x)
        x = self.bn1(x)
        x = F.relu(x)
        x = self.pool1(x)
        x = self.drop1(x)

        x = self.conv2(x)
        x = self.bn2(x)
        x = F.relu(x)
        x = self.pool2(x)
        x = self.drop2(x)

        x = self.conv3(x)
        x = self.bn3(x)
        x = F.relu(x)
        x = self.pool3(x)
        x = self.drop3(x)

        # --- Chuẩn bị cho BiLSTM ---
        # Shape hiện tại: (Batch, Channels=128, NewLength)
        # BiLSTM cần shape: (Batch, SequenceLength, Features/Channels)
        x = x.permute(0, 2, 1)

        # --- Đi qua BiLSTM ---
        out, _ = self.bilstm(x)
        out = self.drop_lstm(out)

        # --- Đi qua Attention ---
        context_vector, _ = self.attention(out)

        # --- Đi qua Dense Layers ---
        out = self.fc1(context_vector)
        out = F.relu(out)
        out = self.drop_fc1(out)

        out = self.fc2(out)
        out = F.relu(out)
        out = self.drop_fc2(out)

        out = self.fc3(out)
        
        # Lưu ý: Không dùng F.softmax ở đây vì hàm nn.CrossEntropyLoss() đã tự động tính Softmax bên trong nó.
        return out

# Khởi tạo mô hình và đẩy lên GPU
model = Paper_CNN_BiLSTM_Attention(num_features=NUM_FEATURES, num_classes=NUM_CLASSES).to(DEVICE)
print(model)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import f1_score, classification_report, confusion_matrix
from tqdm import tqdm

# ==========================================
# 1. THIẾT LẬP HYPERPARAMETERS CHUẨN BÀI BÁO
# ==========================================
# Bài báo sử dụng hàm Loss chuẩn cho phân loại đa lớp
criterion = nn.CrossEntropyLoss()

# Bài báo cấu hình bộ tối ưu hóa Adam với learning rate là 0.001 
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay= 1e-4)

# Sử dụng Scheduler để giảm Learning Rate đi một nửa nếu F1-score đi ngang
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=3, verbose=True)

EPOCHS = 50
EARLY_STOP_PATIENCE = 10 # Cho phép mô hình kiên nhẫn hơn một chút
best_val_f1 = 0.0
patience_counter = 0

print("🚀 BẮT ĐẦU HUẤN LUYỆN MÔ HÌNH 1D-CNN + BiLSTM + ATTENTION...")

# ==========================================
# 2. VÒNG LẶP HUẤN LUYỆN
# ==========================================
for epoch in range(EPOCHS):
    # --- TRAINING ---
    model.train()
    train_loss = 0.0
    
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS} [Train]", leave=False)
    for inputs, labels in loop:
        # Đảm bảo dữ liệu nằm trên GPU
        inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
        
        # Reset gradient
        optimizer.zero_grad()
        
        # Forward pass (Sinh dự đoán)
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        
        # Backward pass (Tính toán đạo hàm)
        loss.backward()
        
        # 🛡️ BẢO HIỂM CHỐNG NỔ LOSS (Gradient Clipping)
        # Giới hạn độ lớn của Gradient không vượt quá 1.0, ngăn chặn Dying ReLU và NaN
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        # Cập nhật trọng số
        optimizer.step()
        
        train_loss += loss.item() * inputs.size(0)
        loop.set_postfix(loss=loss.item())
        
    avg_train_loss = train_loss / len(train_loader.dataset)
    
    # --- VALIDATION ---
    model.eval()
    val_loss = 0.0
    all_preds = []
    all_targets = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(val_loader, desc=f"Epoch {epoch+1}/{EPOCHS} [Val]", leave=False):
            inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
            
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            
            val_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs.data, 1)
            
            all_preds.extend(predicted.cpu().numpy())
            all_targets.extend(labels.cpu().numpy())
            
    avg_val_loss = val_loss / len(val_loader.dataset)
    
    # Đánh giá bằng Macro-F1 (Công bằng cho mọi class)
    val_f1_macro = f1_score(all_targets, all_preds, average='macro')
    
    print(f"Epoch [{epoch+1}/{EPOCHS}] - Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f} | Val Macro-F1: {val_f1_macro:.4f}")
    
    # Điều chỉnh Learning Rate dựa trên F1
    scheduler.step(val_f1_macro)
    
    # --- EARLY STOPPING & BÁO CÁO ---
    if val_f1_macro > best_val_f1:
        best_val_f1 = val_f1_macro
        patience_counter = 0
        
        # Lưu lại tệp trọng số xuất sắc nhất
        torch.save(model.state_dict(), "best_paper_model.pth")
        print(f" -> 🎉 Đã lưu model tốt nhất mới với Val Macro-F1: {best_val_f1:.4f}\n")
    else:
        patience_counter += 1
        if patience_counter >= EARLY_STOP_PATIENCE:
            print(f"\n🛑 [Early Stopping] Mô hình không cải thiện F1 sau {EARLY_STOP_PATIENCE} epochs liên tiếp. Dừng huấn luyện.")
            break

print(f"🎯 Huấn luyện hoàn tất! Best Validation Macro-F1 đạt được: {best_val_f1:.4f}")
```

```python
import torch
import numpy as np
from sklearn.metrics import f1_score, classification_report
from tqdm import tqdm
import optuna

# --- 1. HÀM ÁP DỤNG NGƯỠNG CHO CLASS 10 ---
def apply_stomp_threshold(y_probs, threshold_c10=0.5):
    """
    Hàm ép mô hình phải "tự tin" hơn khi dự đoán Class 10 (STOMP).
    """
    # Lấy ra Top 1 và Top 2 class có xác suất cao nhất
    top2_probs, top2_classes = torch.topk(y_probs, k=2, dim=1)
    
    top1_probs = top2_probs[:, 0]
    top1_classes = top2_classes[:, 0]
    top2_classes = top2_classes[:, 1]
    
    final_preds = top1_classes.clone()
    
    # Kìm hãm Class 10: Nếu đoán là 10 nhưng xác suất < threshold -> Đẩy về Top 2
    mask_c10 = (top1_classes == 10) & (top1_probs < threshold_c10)
    final_preds[mask_c10] = top2_classes[mask_c10]
    
    return final_preds.cpu().numpy()

# --- 2. VÒNG LẶP LẤY XÁC SUẤT TRÊN TẬP VALIDATION ---
# Giả sử bạn đã định nghĩa val_loader và model
all_targets = []
all_probs = []

model.eval()
with torch.no_grad():
    for inputs, labels in tqdm(val_loader, desc="Lấy xác suất trên Val Set"):
        inputs, labels = inputs.to(DEVICE), labels.to(DEVICE)
        outputs = model(inputs)
        
        probs = torch.softmax(outputs, dim=1)
        all_probs.append(probs.cpu())
        all_targets.extend(labels.cpu().numpy())

y_probs_val = torch.cat(all_probs, dim=0)
y_val_true = np.array(all_targets)

y_probs_np = y_probs_val.numpy()
print("\n🚀 BẮT ĐẦU TỐI ƯU HÓA TRỌNG SỐ TOÀN CỤC CHO 11 CLASS BẰNG OPTUNA...")

# 1. Định nghĩa hàm mục tiêu cho Optuna
def objective(trial):
    # Optuna sẽ tự động sinh ra 11 trọng số W trong khoảng từ 0.1 đến 3.0
    weights = [trial.suggest_float(f'w_{i}', 0.1, 3.0) for i in range(11)]
    weights = np.array(weights)
    
    # Nhân xác suất gốc với trọng số
    weighted_probs = y_probs_np * weights
    
    # Lấy class có xác suất mới cao nhất
    y_pred_tuned = np.argmax(weighted_probs, axis=1)
    
    # Trả về Macro F1 để Optuna tối đa hóa
    return f1_score(y_val_true, y_pred_tuned, average='macro')

# 2. Khởi chạy quá trình dò tìm
# Tắt log rác của Optuna cho đỡ rối mắt
optuna.logging.set_verbosity(optuna.logging.WARNING) 
study = optuna.create_study(direction='maximize')

# Chạy 300 vòng thử nghiệm (chỉ mất vài giây vì toàn là phép toán trên ma trận numpy)
study.optimize(objective, n_trials=300, show_progress_bar=True)

# 3. Lấy ra bộ trọng số chiến thắng
best_weights = np.array([study.best_params[f'w_{i}'] for i in range(11)])

print("="*60)
print(f"🎉 Đã chốt được bộ trọng số tối ưu!")
print(f"👉 Trọng số tốt nhất: {np.round(best_weights, 3)}")
print(f"🚀 Macro F1 trên tập Valid đạt: {study.best_value:.4f}")
print("="*60)

# --- 4. IN BÁO CÁO TRÊN TẬP VALID ĐỂ KIỂM CHỨNG ---
weighted_probs_val = y_probs_np * best_weights
final_val_preds = np.argmax(weighted_probs_val, axis=1)

print("\n📊 BÁO CÁO PHÂN LOẠI TRÊN TẬP VALIDATION (SAU KHI ÁP DỤNG TRỌNG SỐ):")
print(classification_report(y_val_true, final_val_preds, digits=4))
```

```python
def custom_macro_f1(y_true, y_pred):
    # Kiểm tra số chiều của mảng dự đoán
    if y_pred.ndim == 2:
        # Nếu là mảng 2D (xác suất), lấy nhãn có xác suất cao nhất
        y_pred_labels = np.argmax(y_pred, axis=1)
    else:
        # Nếu đã là mảng 1D (nhãn trực tiếp), giữ nguyên
        y_pred_labels = y_pred 
        
    # Tính toán Macro F1
    f1 = f1_score(y_true, y_pred_labels, average='macro')
    
    return f1
```
